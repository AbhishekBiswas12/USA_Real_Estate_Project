### Handling Null values

```python
x = [i for i in df.columns if df[f"{i}"].isna().sum()>=1]
print("Percentage of NULL Values: ")
for i in x:
    print(i,": {:.2f}".format((df[f"{i}"].isna().sum()*100/len(df))), "%")
```

```python
df1 = df.drop("sold_date", axis = 1)
df1
```

```python
df1['price'] = df1.price.fillna(df1.price.mean())
df1
```

```python
df1.isna().sum()
```

```python
df1.describe()
```

```python
bed_city = {}

for city in df1['city'].unique():
    bed_city[f"{city}"] = 0
    bed_city[f"{city}"] = df1[df1["city"] == city]['bed'].median()
```
